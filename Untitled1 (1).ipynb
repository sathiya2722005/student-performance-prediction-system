```python
x=df[['Math_Score','English_Score','Science_Score','Study_Hours_Per_Day','Attendance_Percentage']]
y=df['Final_Percentage']
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```

```python
import gradio as gr
import joblib
import numpy as np

# Load trained model
model = joblib.load("student_rf_model.pkl")

# Prediction function
def predict(math, english, science, study_hours, attendance):
    input_data = np.array([[math, english, science, study_hours, attendance]])
    prediction = model.predict(input_data)
    return f"Predicted Final Percentage: {prediction[0]:.2f}%"

# Gradio Interface
app = gr.Interface(
    fn=predict,
    inputs=[
        gr.Number(label="Math Score"),
        gr.Number(label="English Score"),
        gr.Number(label="Science Score"),
        gr.Number(label="Study Hours Per Day"),
        gr.Number(label="Attendance Percentage")
    ],
    outputs="text",
    title="Student Performance Prediction System",
    description="Predicts final academic percentage using Random Forest Regression"
)

app.launch()
```
